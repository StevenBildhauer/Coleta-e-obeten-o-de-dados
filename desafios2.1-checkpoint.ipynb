```python
for jogadores in dados_jogadores.itertuples():
    nome_jogador = jogadores.nome_jogador
    genero = jogadores.genero
    data_nascimento = jogadores.data_nascimento
    d_nascJ = datetime.strptime (data_nascimento,'%d/%m/%Y')
    data_nascimento = d_nascJ.strftime(format="%Y-%m-%d")
    cd_jogo = jogadores.jogo
    cd_pais = jogadores.pais
    cd_vitorias = jogadores.n_vitorias
    cd_derrotas = jogadores.n_derrotas
    total_partidas = jogadores.total_partidas
    
    try:
        query = f"""insert into jogador (nome_jogador, genero, data_nascimento, cod_jogo, 
        cod_pais, n_vitorias, n_derrotas, total_partidas)
                    values ('{nome_jogador}', '{genero}', '{data_nascimento}','{cd_jogo}', '{cd_pais}', '{cd_vitorias}',
                    '{cd_derrotas}', '{total_partidas}')"""
        conn.execute(query)
        print(f"registro inserido com sucesso: {nome_jogador}")
    except Exception as e:
        print(f"não foi possivel inserir o regristo: {nome_jogador}. o erro foi encontrado em: {e}")
```
